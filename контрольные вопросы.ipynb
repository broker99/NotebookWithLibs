```python
import numpy as np
import pandas
import matplotlib.pyplot as mpt 

x = np.arange(-10,10)
y= 1/x

mpt.plot(x,y)
xy=mpt.gca()
mpt.title('гипербола')
xy.spines['left'].set_position(('data',0))
xy.spines['bottom'].set_position(('data',0))
mpt.grid(True)
```
